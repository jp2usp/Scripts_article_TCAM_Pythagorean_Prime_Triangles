```python
from sympy import isprime

def encontrar_triplas_pitagoricas_primitivas(N_limite=200, printNumbers=True,printDiffAnalisis=True):
    """
    Encontra triplas pitagóricas primitivas onde:
    - Hipotenusa é primo da forma 60N+1
    - Cateto menor é primo ímpar > 5Trinca
    - Cateto maior é múltiplo de 60
    """

    hipotenusas_primos = []
    hipotenusas_nao_primos = []
    catetos_menores_primos = []\

    contador_triplas = 0

    print("Triplas Pitagóricas Primitivas Especiais:")
    print("=" * 60)
    print("Calculations in process\n")
    for i in range(1, N_limite):
        # Calcula candidato a hipotenusa
        q = 2 * i**2 - 2 * i + 1

        # Filtro 1: deve ser da forma 60N+1
        if q % 60 != 1:
            continue

        # Filtro 2: deve ser primo
        if not isprime(q):
            hipotenusas_nao_primos.append(q)
            continue

        # Calcula cateto menor (p = 2i-1)
        p = 2 * i - 1

        # Filtro 3: cateto menor deve ser primo
        if not isprime(p):
            hipotenusas_nao_primos.append(q)
            continue

        # Verifica propriedade pitagórica
        b = q - 1  # cateto maior
        n = b // 60  # N na fórmula 60N

        if q**2 == p**2 + b**2:
            contador_triplas += 1
            hipotenusas_primos.append(q)
            catetos_menores_primos.append(p)

            if printNumbers==True:
              print(f"{contador_triplas:3d}.) c={q:4d} | a={p:4d} | b={b:4d}=60×{n:2d}")

    print("=" * 60)

    if printDiffAnalisis==True:

      # Análise de diferenças entre catetos menores primos
      if len(catetos_menores_primos) >= 2:
          diferencas = [catetos_menores_primos[i+1] - catetos_menores_primos[i]
                      for i in range(len(catetos_menores_primos)-1)]

          maior_diferenca = max(diferencas)
          indice_maior = diferencas.index(maior_diferenca)

          print(f"\nAnálise de Diferenças entre Catetos Menores Primos:")
          print(f"Quantidade de triplas encontradas: {contador_triplas}")
          print(f"Maior diferença: {maior_diferenca}")
          print(f"Ocorre entre {catetos_menores_primos[indice_maior]} e {catetos_menores_primos[indice_maior+1]}")

      print(f"\nHipotenusas 60N+1 que não formam triplas: {len(hipotenusas_nao_primos)}")
    print("Calculations terminated\n")
    return hipotenusas_primos, hipotenusas_nao_primos, catetos_menores_primos


hipotenusas_primos, hipotenusas_nao_primos, catetos_menores = \
        encontrar_triplas_pitagoricas_primitivas(200,printNumbers=False,printDiffAnalisis=False)
```

```python
from sympy.ntheory import nextprime
from sympy.ntheory import isprime
from typing import List, Tuple

def encontrar_triplos_primos_pitagoricos(primo1_inicio: int, Ntrincas: int) -> Tuple[List[int], List[int], List[int]]:
    """
    Busca Triângulos Primos de Pitágoras com o cateto (primo1) e hipotenusa (q)
    sendo ímpares, utilizando a lógica do seu algoritmo.

    Args:
        primo1_inicio: O primeiro valor primo a ser testado como cateto.
        Ntrincas: A quantidade desejada de Triângulos Primos de Pitágoras (primo_hipotenusa).

    Returns:
        Uma tupla contendo as listas: (primo_hipotenusa, Notprimo_hipotenusa, primo_cateto).
    """

    primo1 = primo1_inicio
    primo_cateto: List[int] = []
    primo_hipotenusa: List[int] = []
    Notprimo_hipotenusa: List[int] = []

    # O loop continua até que a quantidade de hipotenusas primas (primo_hipotenusa)
    # atinja a quantidade Ntrincas desejada.
    while len(primo_hipotenusa) < Ntrincas:
        # 1. Cálculos auxiliares
        x = int((primo1 + 1) / 2)
        aux = int(x**2)

        # 2. Encontra o próximo primo para a próxima iteração
        primo2 = nextprime(primo1)

        # 3. Verifica a condição do algoritmo: (a+b+c)/2 - a = 30k
        # onde a=primo1, b=60N, c=q (a, b e c são lados do triângulo)
        # O cálculo simplificado do código é: (aux - x) % 30 == 0
        if ((aux - x) % 30 == 0):

            # Se a condição for atendida, o primo1 é um cateto válido
            primo_cateto.append(primo1)

            # Cálculo da Hipotenusa q: 2*x^2 - primo1
            q = int(2 * aux - primo1)

            # Verifica se a hipotenusa q é prima
            if isprime(q):
                primo_hipotenusa.append(q)
            else:
                Notprimo_hipotenusa.append(q)

        # Avança para o próximo primo para a próxima iteração
        primo1 = primo2

    return primo_hipotenusa, Notprimo_hipotenusa, primo_cateto

# ==============================================================================
# EXEMPLO DE USO DA FUNÇÃO
# ==============================================================================

primo1_inicio = 11
Ntrincas_desejadas = 100

print(f"Buscando {Ntrincas_desejadas} Triângulos Primos de Pitágoras, começando com primo1 = {primo1_inicio}...")

# Chamada da função
hip_primas, hip_nao_primas, catetos_primos = encontrar_triplos_primos_pitagoricos(primo1_inicio, Ntrincas_desejadas)

print("\n------------------------------------------------------")
print("RESULTADOS:")
print("------------------------------------------------------")

# Imprime as listas retornadas
print("\nLista de Hipotenusas Primas (q):")
print(hip_primas)
print(f"Primos Hipotenusas: {len(hip_primas)} ocorrências")

print("\nLista de Hipotenusas Não-Primas (q):")
print(hip_nao_primas)
print(f"Não-Primo Hipotenusa: {len(hip_nao_primas)} ocorrências")

print("\nLista de Catetos Primos (primo1):")
print(catetos_primos)
print(f"Primos Catetos: {len(catetos_primos)} ocorrências")
```

```python
import math
from sympy.ntheory import factorint

def prime_factorization(n):
    """
    Decomposes a natural number into its prime factors.

    Args:
        n (int): A natural number (integer greater than 0).

    Returns:
        dict: A dictionary where keys are prime factors and values are their exponents.
              Returns an empty dictionary if n is 1, and raises ValueError for non-natural numbers.
    """
    if not isinstance(n, int) or n < 1:
        raise ValueError("Input must be a natural number (integer greater than 0).")
    if n == 1:
        return {}
    return factorint(n)

# Example usage:
number_to_factorize = 240
factors = prime_factorization(number_to_factorize)
print(f"The prime factorization of {number_to_factorize} is: {factors}")

number_to_factorize = 997
factors = prime_factorization(number_to_factorize)
print(f"The prime factorization of {number_to_factorize} is: {factors}")

number_to_factorize = 1
factors = prime_factorization(number_to_factorize)
print(f"The prime factorization of {number_to_factorize} is: {factors}")

for q in hip_nao_primas:
  p=int(math.sqrt(2*q-1))
  print(f"The prime factorization of q={q} is {prime_factorization(q)} and p={p}")
```

```python
# Função para razão hip_primas/hip_nao_primas
from sympy.ntheory import nextprime
from sympy.ntheory import isprime
from typing import Dict, List, Any

def calcular_totais_cumulativos(primo1_inicio: int, Ntrincas_max: int) -> Dict[str, List[int]]:
    """
    Calcula e retorna os totais acumulados (len) de primo_cateto, hip_nao_primas,
    etc., para cada Triângulo Primo de Pitágoras encontrado, de 1 até Ntrincas_max.
    """

    primo1 = primo1_inicio

    # Listas para armazenar o estado cumulativo a cada passo
    qnt_hip_primas_acumulado: List[int] = []
    qnt_hip_nao_primas_acumulado: List[int] = []
    qnt_catetos_acumulado: List[int] = []

    # Contadores temporários
    count_hip_primas = 0
    count_hip_nao_primas = 0
    count_catetos = 0

    print(f"Buscando até {Ntrincas_max} Hipotenusas Primas...")

    while count_hip_primas < Ntrincas_max:
        x = int((primo1 + 1) / 2)
        aux = int(x**2)
        primo2 = nextprime(primo1)

        if ((aux - x) % 30 == 0):
            count_catetos += 1

            q = int(2 * aux - primo1)

            if isprime(q):
                count_hip_primas += 1

                # Grava o estado APÓS encontrar a n-ésima hipotenusa prima
                qnt_hip_primas_acumulado.append(count_hip_primas)
                qnt_hip_nao_primas_acumulado.append(count_hip_nao_primas)
                qnt_catetos_acumulado.append(count_catetos)
            else:
                count_hip_nao_primas += 1

        primo1 = primo2
    print(f"obtido {Ntrincas_max} Hipotenusas Primas...")
    return {
        "primos_pitagoras_N": [i + 1 for i in range(Ntrincas_max)],
        "total_hip_primas": qnt_hip_primas_acumulado,
        "total_hip_nao_primas": qnt_hip_nao_primas_acumulado,
        "total_catetos_primos": qnt_catetos_acumulado
    }

## Exemplo de Execução e Análise

primo1_inicio = 11
Ntrincas_desejadas = 100 # Queremos os totais até a 100ª Hipotenusa Prima

resultados_cumulativos = calcular_totais_cumulativos(primo1_inicio, Ntrincas_desejadas)

# As colunas contêm os totais no momento em que a n-ésima hipotenusa prima foi encontrada.

# Para análise, vamos usar a biblioteca Pandas para visualizar melhor
import pandas as pd

df_resultados = pd.DataFrame(resultados_cumulativos)

# Renomear a coluna 'primos_pitagoras_N' para o índice (1 a 100)
df_resultados.set_index("primos_pitagoras_N", inplace=True)
df_resultados.index.name = "N-ésima Hipotenusa Prima Encontrada"

# Exibir as 10 primeiras linhas do DataFrame
print("\nDataFrame de Totais Cumulativos (Primeiras 10 linhas):")
print(df_resultados.head(10))

# Exibir as últimas 5 linhas (Total final)
print("\nTotais Finais (N=100):")
print(df_resultados.tail(5))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sympy.ntheory import nextprime
from sympy.ntheory import isprime
from typing import Dict, List, Any

# ==============================================================================
# 2. Geração de Dados e Cálculo da Razão
# ==============================================================================

primo1_inicio = 11
Ntrincas_desejadas = 100000# Usando 1000 para um gráfico mais robusto

resultados_cumulativos = calcular_totais_cumulativos(primo1_inicio, Ntrincas_desejadas)
df = pd.DataFrame(resultados_cumulativos)
```
